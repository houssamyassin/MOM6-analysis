### Plot sea surface height

```python
%%time
# -------------------------------------------------------------------
SSH_means         = []   # time-mean bias-corrected SSH fields
res_funcs         = []   # time-mean resolution functions
# -------------------------------------------------------------------
# Observational reference
obs_path = '/glade/work/gmarques/cesm/tx1_4/ssh/adt_mean_aviso_to_tx1_4.nc'
obs  = xr.open_dataset(obs_path)
sims[0].open_single_file(sim.native_path,date='0055-01')
sims[0].open_grid()
obs  = obs.rename({'ny':'yh', 'nx':'xh'}).assign_coords({
         'geolon': sims[0].grid['geolon'],
         'geolat': sims[0].grid['geolat'],
      })
obs_bias = obs.adt_mean.mean().compute()      # a single scalar (≈ 0.024 m)

# -------------------------------------------------------------------
for i, sim in enumerate(sims):
    # ── open data
    sim.open_files_by_year(sim.ocn_data_path, sim.native_end, start_year, end_year)
    sim.open_grid()
    ds = sim.ds.sel(time=slice(start_date, end_date))

    # ── global-mean bias and bias-corrected SSH ---------------------
    model_bias = ds.SSH.mean(dim=['time', 'yh', 'xh']).compute()
    Δ          = model_bias - obs_bias            # scalar offset (model – obs)
    SSH_bc     = ds.SSH - Δ                       # bias-corrected absolute SSH

    # ── time-mean fields (for optional plotting) --------------------
    SSH_mean_bc = SSH_bc.mean('time').compute()
    SSH_means.append(SSH_mean_bc)

    Rd_dx       = ds.Rd_dx.mean('time').compute()
    res_funcs.append(resolution(Rd_dx, alpha_res, p_res))



```

```python
diffs = []
for i in range(1,num_sims):
    diff = SSH_means[i].fillna(0)-SSH_means[0].fillna(0)
    diff = diff.assign_coords({'geolon': sims[0].grid['geolon'],
                       'geolat': sims[0].grid['geolat']})
    diff = diff.where(diff != 0, np.nan)

    diffs.append(diff)
```

```python
import matplotlib.colors as mcolors
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
cmap = cm.RdYlBu
vmin, vmax = -0.5, 0.5
n_steps = 8  # how many discrete bands you want
levels = np.linspace(vmin, vmax, n_steps + 1)   # e.g. [0.0,0.25,0.5,…,2.0]

# create a BoundaryNorm so each interval maps to one color
norm = mcolors.BoundaryNorm(boundaries=levels, ncolors=cm.RdYlBu.N, clip=True)

for sim, R, Rd in zip(sims[1:], res_funcs[1:], diffs):
    title = sim.short_name + " - REF"
    # pass norm instead of vmin/vmax
    mappable = sim.plot_global_discrete(Rd,
                                        cmap=cmap,
                                        levels=levels,
                                        cbar_label= r'',
                                        title=title,
                                        figsize=(12,8),
                                        R=R,
                                        R_threshold=0.5,
                                        contour_color='k', 
                                        contour_linewidth=1.5, 
                                        contour_linestyle='-')
    
    ax = plt.gca()
    
    # set up graticules
    lon_ticks = np.arange(-180, 181, 60)
    lat_ticks = np.arange(-90, 91, 30)
    ax.set_xticks(lon_ticks, crs=ccrs.PlateCarree())
    ax.set_yticks(lat_ticks, crs=ccrs.PlateCarree())
    ax.xaxis.set_major_formatter(LongitudeFormatter())
    ax.yaxis.set_major_formatter(LatitudeFormatter())
    ax.gridlines(draw_labels=False, linestyle='--', color='gray', alpha=0.5)
    plt.savefig(save_path+"SSH_"+title+'.png',dpi=300)
```
